```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import PandasTools
from rdkit.Chem import AllChem
```

```python
df=pd.read_excel('DF_1225x6_12-19-2022.xlsx', index_col='PubChem CID')
#codes=df.iloc[:,0]
#smi=df.iloc[:,2]
```

```python
molecs=[]
errligs=[]
errmols=[]
for i,j in zip(list(df['InChI']),list(df.index)):
    m=Chem.MolFromInchi(i)
#    try:
#        lig1=Chem.AddHs(m)
#    except:
#        errligs.append(j) 
    try:
        #AllChem.EmbedMolecule(lig1)
        #AllChem.UFFOptimizeMolecule(lig1, maxIters=20000, ignoreInterfragInteractions=False)
    #Chem.rdForceFieldHelpers.MMFFSanitizeMolecule(lig1)
        lig1 = m
        lig1=Chem.AddHs(m)
        AllChem.EmbedMolecule(lig1)
        Chem.rdForceFieldHelpers.MMFFOptimizeMolecule(lig1,mmffVariant='MMFF94s',maxIters=500)
        molecs.append(lig1)
        Chem.rdmolfiles.MolToPDBFile(lig1, str(j)+"_MMFF94s.pdb")
        print(j,' - Done!')
    except:
        errmols.append(j)
        print('error...', j)
print(type(molecs[2]),len(molecs))
```
